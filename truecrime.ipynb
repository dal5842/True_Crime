```python
print(doc.text)
```

```python
# Assign your list to a variable
proper_nouns = [prop.text for prop in doc if prop.pos_ == "PROPN"]

# Now write to a text file
with open("proper_nouns.txt", "w", encoding="utf-8") as f:
    for word in proper_nouns:
        f.write(word + "\n")
```

```python
entities = [(ent.text, ent.label_) for ent in doc.ents]

with open("named_entities.txt", "w", encoding="utf-8") as f:
    for text, label in entities:
        f.write(f"{text} ({label})\n")
```

```python
verbs = [token.lemma_ for token in doc if token.pos_ == "VERB"]

with open("verbs.txt", "w", encoding="utf-8") as f:
    for verb in verbs:
        f.write(verb + "\n")
```
